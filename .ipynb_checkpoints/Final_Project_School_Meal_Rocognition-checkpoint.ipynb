```python
import os
import cv2
import glob 
import numpy as np
import pandas as pd
import skimage.io as io 
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow.keras.layers as layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
from tensorflow.keras.applications import InceptionResNetV2
from tensorflow.keras.applications import NASNetLarge
from tensorflow.keras.applications import EfficientNetB0
from tqdm import tqdm,trange
from keras.callbacks import ModelCheckpoint
```

```python
# 確認壓縮檔路徑
filepath='/Users/boboloiono/Documents/電腦視覺與深度學習/Final Project/'
train_path = '/Users/boboloiono/Documents/電腦視覺與深度學習/Final Project/train/'
test_path = '/Users/boboloiono/Documents/電腦視覺與深度學習/Final Project/test/'
```

## Image Preprocessing

```python
imgSize=(100,100)
imgShape=(100,100,3)
batchSize = 64

# create image data augmentation generator
trainDataGenerator = ImageDataGenerator(
                      zoom_range = 0.3,
                      rotation_range = 20,
                      width_shift_range=0.3,
                      height_shift_range=0.3,
                      horizontal_flip = True,
                      validation_split=0.3)

trainBatch = trainDataGenerator.flow_from_directory(
    directory = train_path,
    target_size = imgSize,
    class_mode = 'categorical',
    shuffle = True,
    batch_size = batchSize,
    subset='training'
)
```

```python
class_names = []
for cls, idx in trainBatch.class_indices.items():
    class_names.append(cls)
    print('Class #{} = {}'.format(idx, cls))
```

```python
plt.figure(figsize = (20, 10))
for i in range(50):
    food_path = os.path.join(train_path, class_names[i], '*')
    food_path = glob.glob(food_path)
    image = io.imread(food_path[5])
    image = cv2.resize(image,(120,120))
    plt.subplot(5,10,i+1)
    plt.imshow(image)
    plt.title(class_names[i])
```

```python
from numpy import expand_dims

# expand dimension to one sample
samples = expand_dims(image, 0)

it = trainDataGenerator.flow(samples, batch_size=1)
fig, im = plt.subplots(nrows=1, ncols=9, figsize=(15,15))

# generate batch of images
for i in range(9):
    image = next(it)[0].astype('uint8')
    im[i].imshow(image)
```

```python
imgSize=(224, 224)
imgShape=(224, 224, 3)
batchSize = 64

# create image data augmentation generator
trainDataGenerator = ImageDataGenerator(rescale=1/255,
                      zoom_range = 0.3,
                      rotation_range = 20,
                      width_shift_range=0.3,
                      height_shift_range=0.3,
                      horizontal_flip = True,
                      validation_split=0.2)

valDataGenerator = ImageDataGenerator(rescale=1/255,validation_split=0.2)

trainBatch = trainDataGenerator.flow_from_directory(
    directory = train_path,
    target_size = imgSize,
    class_mode = 'categorical',
    shuffle = True,
    batch_size = batchSize,
    subset='training'
)

valBatch = valDataGenerator.flow_from_directory(
    directory = train_path,
    target_size = imgSize,
    class_mode = 'categorical',
    shuffle = False,
    batch_size = batchSize,
    subset='validation'
)

# test generator
testDataGenerator = ImageDataGenerator(rescale=1/255)

testBatch = testDataGenerator.flow_from_directory(
    test_path,
    class_mode='categorical',
    batch_size = batchSize,
    target_size= imgSize,
    shuffle = False
)
```

```python
base_model = tf.keras.applications.InceptionResNetV2(
    include_top = False,
    input_shape=imgShape,
)

x = base_model.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(512, activation = 'relu')(x)
x = layers.Dropout(0.2)(x)
output_layer = layers.Dense(units = 50, activation = 'softmax')(x)

# 構建我們需要訓練的完整模型
model = tf.keras.Model(inputs = base_model.input, outputs = output_layer)

for layer in base_model.layers:
    layer.trainable = False
```

```python
model.summary()
```

```python
# ---------------------------2.設置callbacks----------------------------
filepath="weights-improvement-{epoch:02d}-{val_accuracy:.2f}.h5"
checkpointer = ModelCheckpoint(filepath, verbose=1, monitor='val_accuracy',save_best_only=True, mode='max')
```

```python
model.load_weights('weights-improvement-10-0.67.h5')
# 在新的數據集上訓練幾代
model.compile(optimizer = Adam(learning_rate=0.001, decay=0.01), 
              loss='categorical_crossentropy',metrics=['accuracy',
              tf.keras.metrics.TopKCategoricalAccuracy(k=5, name='top5_categorical_accuracy', dtype=None)])
```

```python
# -----------------------------4.開始訓練模型------------------------------
# 重新訓練權重
history = model.fit_generator(trainBatch,
                   epochs=8, verbose=1,
                   steps_per_epoch=trainBatch.samples//batchSize,
                   validation_data=valBatch,
                   validation_steps=valBatch.samples//batchSize,
                   callbacks=checkpointer)
```

```python
# -----------------------5.儲存模型、紀錄學習曲線------------------------
# 儲存模型
# model.save('./InceptionV4_retrained_v2.h5')
# print('已儲存InceptionV2_retrained_v2.h5')

# 畫出acc學習曲線
acc = history.history['accuracy']
epochs = range(1, len(acc) + 1)
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend(loc='lower right')
plt.grid()
# 儲存acc學習曲線
plt.savefig('./acc.png')
plt.show()

# 畫出loss學習曲線
loss = history.history['loss']
val_loss = history.history['val_loss']
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.legend(loc='upper right')
plt.grid()
# 儲存loss學習曲線
plt.savefig('loss.png')
plt.show()
```

```python
# -------------------------6.驗證模型準確度--------------------------
# 以vali資料夾驗證模型準確度
test_loss, test_acc = model.evaluate_generator(testBatch,
                            steps=testBatch.samples//batchSize,
                            verbose=1)
print('test acc:', test_acc)
print('test loss:', test_loss)
```

## Classification Report

```python
filenames = testBatch.filenames
nb_samples = len(filenames)

# Making predictions
predict = model.predict(testBatch, steps = nb_samples)
predicted_classes = predict.argmax(axis=-1)

# Confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sn

print(classification_report(testBatch.classes, predicted_classes))
result = confusion_matrix(testBatch.classes, predicted_classes)
fig = plt.figure(figsize=(30, 30))
sn.heatmap(result, annot=True, cmap="OrRd", fmt='g')
```
